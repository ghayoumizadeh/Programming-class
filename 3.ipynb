# **📌 دستورات کلیدی و پرکاربرد در پایتون**

پایتون دارای چندین دستور کلیدی است که در بسیاری از برنامه‌ها مورد استفاده قرار می‌گیرد. این دستورات برای بررسی نوع داده‌ها، طول، ویژگی‌ها و سایر مشخصات متغیرها بسیار مفید هستند.

---

## **🔹 ۱. بررسی نوع داده با `type()`**
`type()` برای بررسی **نوع داده** یک متغیر یا مقدار استفاده می‌شود.

</div>

```python
x = 10
print(type(x))  # خروجی: <class 'int'>

y = "Hello"
print(type(y))  # خروجی: <class 'str'>

z = [1, 2, 3]
print(type(z))  # خروجی: <class 'list'>
```

<div dir="rtl" style="text-align: right;">

## **🔹 ۲. بررسی طول داده با `len()`**
`len()` تعداد عناصر در یک **رشته، لیست، تاپل، دیکشنری و سایر ساختارهای داده‌ای** را برمی‌گرداند.

</div>

```python
text = "Python"
print(len(text))  # خروجی: 6

numbers = [1, 2, 3, 4, 5]
print(len(numbers))  # خروجی: 5

dictionary = {"name": "Ali", "age": 25}
print(len(dictionary))  # خروجی: 2 (تعداد کلیدها)
```

# **رشته‌ها (String) در پایتون**

### **📝 تعریف رشته در پایتون**
در پایتون، **رشته (String)** یک **داده متنی** است که بین **علامت‌های نقل‌قول** (`' '` یا `" "`) قرار می‌گیرد. رشته‌ها می‌توانند شامل **حروف، اعداد، و نمادها** باشند.

### **📌 نحوه تعریف رشته:**
- می‌توان رشته را داخل `"` یا `'` قرار داد.
- برای رشته‌های چندخطی از `"""` یا `'''` استفاده می‌شود.

</div>

```python
# تعریف رشته‌های تک‌خطی
text1 = "سلام دنیا!"
text2 = 'Python عالی است.'

# تعریف رشته چندخطی
text3 = """این یک متن 
چندخطی در پایتون است."""
```

<div dir="rtl" style="text-align: right;">

---

### **✂️ دسترسی به کاراکترهای رشته**
هر کاراکتر در یک رشته **دارای یک اندیس (index)** است.  
**اندیس‌ها از `۰` شروع می‌شوند.**  
همچنین می‌توان از **اندیس منفی** استفاده کرد که از آخر رشته شمارش می‌شود.

</div>

```python
text = "Python"

print(text[0])   # خروجی: P
print(text[-1])  # خروجی: n (آخرین کاراکتر)
```

<div dir="rtl" style="text-align: right;">

---

### **✂️ برش دادن (Slicing) در رشته**
برای دسترسی به بخشی از رشته، می‌توان از **برش (Slicing)** استفاده کرد.  
فرمت کلی:
```python
string[start:end:step]
```
- `start`: موقعیت شروع (پیش‌فرض `0`)
- `end`: موقعیت پایان (تا قبل از این ایندکس)
- `step`: تعداد گام‌ها (پیش‌فرض `1`)

</div>

```python
text = "Python"

print(text[0:4])   # خروجی: Pyth
print(text[::2])   # خروجی: Pto (پرش ۲تایی)
print(text[::-1])  # خروجی: nohtyP (معکوس کردن رشته)
```

<div dir="rtl" style="text-align: right;">

---

### **🔄 عملیات روی رشته‌ها**
در پایتون، رشته‌ها **غیرقابل تغییر (Immutable)** هستند، اما می‌توان روی آن‌ها عملیات مختلفی انجام داد.

</div>

```python
text = "Python"

print(text.upper())    # تبدیل به حروف بزرگ → خروجی: PYTHON
print(text.lower())    # تبدیل به حروف کوچک → خروجی: python
print(text.replace("P", "J"))  # جایگزینی → خروجی: Jython
print(len(text))       # طول رشته → خروجی: 6
```

<div dir="rtl" style="text-align: right;">

---

### **📌 بررسی عضویت در رشته**
برای بررسی وجود یک زیررشته در داخل رشته، می‌توان از `in` استفاده کرد.

</div>

```python
text = "سلام به دنیای پایتون!"

print("پایتون" in text)  # خروجی: True
print("جاوا" not in text)  # خروجی: True
```

<div dir="rtl" style="text-align: right;">

---

### **🔗 اتصال رشته‌ها (Concatenation)**
با استفاده از `+` می‌توان چند رشته را به هم متصل کرد.

</div>

```python
first = "سلام"
second = "دنیا"
message = first + " " + second + "!"
print(message)  # خروجی: سلام دنیا!
```

<div dir="rtl" style="text-align: right;">

---

### **📌 قالب‌بندی رشته‌ها (f-strings)**
در **پایتون ۳.۶+** می‌توان از **f-strings** برای قالب‌بندی رشته‌ها استفاده کرد.

</div>

```python
name = "علی"
age = ۲۵

print(f"نام: {name}, سن: {age}")  # خروجی: نام: علی, سن: ۲۵
```

<div dir="rtl" style="text-align: right;">

---

### **📌 تبدیل انواع داده به رشته**
می‌توان با استفاده از `str()` انواع داده مانند عدد را به رشته تبدیل کرد.

</div>

```python
number = 100
text = str(number)

print(text)      # خروجی: '100'
print(type(text))  # خروجی: <class 'str'>
```
