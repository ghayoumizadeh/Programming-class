# **📌 حلقه `while` در پایتون**

### **📝 حلقه `while` چیست؟**
در پایتون، **حلقه `while`** برای **اجرای مداوم یک مجموعه از دستورات تا زمانی که یک شرط برقرار باشد** استفاده می‌شود.  
📌 **اگر شرط `True` باشد، حلقه اجرا می‌شود و زمانی که `False` شد، حلقه متوقف خواهد شد.**  
🎯 **از `while` معمولاً برای اجرای بی‌نهایت یا زمانی که تعداد تکرارها مشخص نیست، استفاده می‌شود.**

---

## **🔹 ۱. ساختار کلی حلقه `while` در پایتون**
📌 **ساختار کلی:**
</div>

```python
while شرط:
    # دستورات اجرا می‌شوند تا زمانی که شرط True باشد
```

<div dir="rtl" style="text-align: right;">

🔹 **شرط بررسی می‌شود:** اگر مقدار آن `True` باشد، کد داخل حلقه اجرا می‌شود.  
🔹 **پس از هر تکرار، مقدار شرط بررسی می‌شود:** اگر هنوز `True` باشد، حلقه ادامه می‌یابد، در غیر این صورت متوقف می‌شود.  

---

## **🔹 ۲. مثال ساده از `while`**
📌 **حلقه زیر اعداد ۱ تا ۵ را چاپ می‌کند.**  
🔹 **متغیر `i` در ابتدا مقدار `۱` دارد و در هر تکرار `۱` واحد افزایش می‌یابد.**

</div>

```python
i = 1

while i <= 5:
    print(f"عدد: {i}")
    i += 1  # افزایش مقدار i برای جلوگیری از حلقه بی‌نهایت
```

<div dir="rtl" style="text-align: right;">

📌 **توضیح:**  
- مقدار اولیه `i = 1` است.  
- حلقه تا زمانی که `i <= 5` باشد، اجرا می‌شود.  
- در هر تکرار، مقدار `i` افزایش می‌یابد (`i += 1`).  
- اگر این افزایش را نگذاریم، حلقه **بی‌نهایت** اجرا می‌شود.

---

## **🔹 ۳. استفاده از `while` برای گرفتن ورودی**
📌 **این مثال از `while` برای دریافت ورودی از کاربر استفاده می‌کند تا زمانی که مقدار معتبر وارد شود.**

</div>

```python
password = ""

while password != "1234":
    password = input("رمز عبور را وارد کنید: ")

print("رمز صحیح است، ورود موفقیت‌آمیز!")
```

<div dir="rtl" style="text-align: right;">

📌 **توضیح:**  
- تا زمانی که کاربر مقدار `"1234"` را وارد نکند، حلقه اجرا خواهد شد.  
- به محض ورود مقدار صحیح، پیام `"رمز صحیح است"` نمایش داده شده و حلقه متوقف می‌شود.

---

## **🔹 ۴. استفاده از `break` برای خروج از `while`**
📌 **اگر بخواهیم در میانه حلقه متوقف شویم، می‌توان از `break` استفاده کرد.**  
🔹 در این مثال، اگر عدد ورودی **`0`** باشد، حلقه متوقف می‌شود.

</div>

```python
while True:
    num = int(input("یک عدد وارد کنید (0 برای خروج): "))
    
    if num == 0:
        print("خروج از حلقه!")
        break  # خروج از حلقه
    
    print(f"شما عدد {num} را وارد کردید.")
```

<div dir="rtl" style="text-align: right;">

📌 **توضیح:**  
- حلقه **بی‌نهایت (`while True`)** اجرا می‌شود.  
- اگر عدد **`0`** وارد شود، `break` باعث توقف حلقه می‌شود.  
- در غیر این صورت، مقدار عدد چاپ خواهد شد.

---

## **🔹 ۵. استفاده از `continue` برای رد کردن یک تکرار**
📌 **`continue` باعث می‌شود که ادامه کد اجرا نشود و به ابتدای حلقه برگردد.**  
🔹 در این مثال، اگر عدد **زوج** باشد، از چاپ آن صرف‌نظر می‌شود.

</div>

```python
i = 0

while i < 10:
    i += 1
    
    if i % 2 == 0:
        continue  # از این مرحله رد می‌شود و حلقه ادامه پیدا می‌کند
    
    print(f"عدد فرد: {i}")
```

<div dir="rtl" style="text-align: right;">

📌 **توضیح:**  
- مقدار `i` در هر تکرار **افزایش** می‌یابد.  
- اگر مقدار `i` **زوج** باشد، `continue` اجرا شده و چاپ انجام نمی‌شود.  
- در نهایت، فقط اعداد **فرد** نمایش داده می‌شوند.

---

## **🔹 ۶. اجرای حلقه `while` با `else`**
📌 **در پایتون، `else` می‌تواند بعد از `while` استفاده شود و زمانی اجرا خواهد شد که حلقه به‌طور طبیعی (بدون `break`) متوقف شود.**

</div>

```python
i = 1

while i <= 3:
    print(f"مرحله {i}")
    i += 1
else:
    print("حلقه به پایان رسید!")
```

<div dir="rtl" style="text-align: right;">

📌 **توضیح:**  
- حلقه تا مقدار `i = 3` اجرا شده و سپس متوقف می‌شود.  
- چون `break` استفاده نشده، **بخش `else` اجرا می‌شود**.

---

## **🔹 ۷. حلقه `while` بی‌نهایت**
📌 **اگر شرط حلقه همیشه `True` باشد، حلقه **بی‌نهایت** خواهد شد.**  
🔹 در این مثال، تا زمانی که کاربر `exit` را وارد نکند، حلقه ادامه دارد.

</div>

```python
while True:
    command = input("دستور خود را وارد کنید (exit برای خروج): ")
    
    if command == "exit":
        print("برنامه متوقف شد.")
        break
    
    print(f"دستور شما: {command}")
```

<div dir="rtl" style="text-align: right;">

📌 **توضیح:**  
- `while True` یک **حلقه بی‌نهایت** ایجاد می‌کند.  
- اگر کاربر `"exit"` وارد کند، `break` باعث خروج از حلقه می‌شود.  
- در غیر این صورت، مقدار ورودی نمایش داده خواهد شد.
