# **📌 کار با اعداد (`int` و `float`) در پایتون**

پایتون دارای دو نوع عددی اصلی است:
- **عدد صحیح (`int`)**: اعداد بدون قسمت اعشاری، مانند `1، 100، -50`.
- **عدد اعشاری (`float`)**: اعداد با قسمت اعشاری، مانند `3.14، -2.5، 10.0`.

---

## **🔹 ۱. تعریف `int` و `float` در پایتون**
در پایتون، اعداد صحیح (`int`) و اعشاری (`float`) را می‌توان به‌سادگی تعریف کرد.

</div>

```python
# تعریف عدد صحیح (int)
x = 10
y = -5

# تعریف عدد اعشاری (float)
a = 3.14
b = -2.75

# چاپ نوع متغیرها
print(type(x))  # خروجی: <class 'int'>
print(type(a))  # خروجی: <class 'float'>
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۲. عملیات ریاضی روی اعداد**
در پایتون می‌توان عملیات ریاضی مختلفی روی `int` و `float` انجام داد.

| **عملیات**   | **نماد** | **مثال**           | **خروجی** |
|-------------|---------|----------------|----------|
| **جمع**     | `+`     | `10 + 5`       | `15`     |
| **تفریق**   | `-`     | `10 - 3`       | `7`      |
| **ضرب**     | `*`     | `4 * 2`        | `8`      |
| **تقسیم**   | `/`     | `10 / 3`       | `3.333`  |
| **تقسیم صحیح** | `//`  | `10 // 3`      | `3`      |
| **باقی‌مانده** | `%`   | `10 % 3`       | `1`      |
| **توان**    | `**`    | `2 ** 3`       | `8`      |

---

## **🔹 ۳. مثال‌هایی از عملیات ریاضی**

</div>

```python
# عملیات ریاضی روی اعداد صحیح و اعشاری
x = 10
y = 3

print(x + y)   # جمع → خروجی: 13
print(x - y)   # تفریق → خروجی: 7
print(x * y)   # ضرب → خروجی: 30
print(x / y)   # تقسیم → خروجی: 3.333...
print(x // y)  # تقسیم صحیح → خروجی: 3
print(x % y)   # باقیمانده → خروجی: 1
print(x ** y)  # توان → خروجی: 1000
```

<div dir="rtl" style="text-align: right;">

## **🔹 ۴. تبدیل `int` و `float` به یکدیگر**
گاهی اوقات لازم است بین `int` و `float` تبدیل انجام دهیم.

</div>

```python
# تبدیل int به float
x = 10
x_float = float(x)
print(x_float)  # خروجی: 10.0

# تبدیل float به int (مقدار اعشار حذف می‌شود)
y = 3.75
y_int = int(y)
print(y_int)  # خروجی: 3
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۵. مقداردهی ترکیبی (جمع و ضرب با مقدار قبلی)**
می‌توان مقدار یک متغیر را با استفاده از عملگرهای `+=` و `*=` تغییر داد.

</div>

```python
x = 5
x += 3  # معادل x = x + 3
print(x)  # خروجی: 8

y = 4
y *= 2  # معادل y = y * 2
print(y)  # خروجی: 8
```

✅ `int` و `float` دو نوع داده عددی مهم در پایتون هستند.  
✅ عملیات ریاضی مانند `+ - * /` به‌راحتی روی این اعداد قابل انجام است.  
✅ `//` برای تقسیم صحیح و `%` برای باقیمانده استفاده می‌شود.  
✅ می‌توان `int` را به `float` تبدیل کرد و بالعکس.  
🎯 **یادگیری کار با اعداد در پایتون برای توسعه برنامه‌های محاسباتی و علمی ضروری است.** 🚀

# **📌 اولویت عملگرهای ریاضی در پایتون**

پایتون هنگام محاسبات ریاضی، عملگرها را بر اساس **اولویت مشخصی** اجرا می‌کند.  
**عملگرهایی که اولویت بالاتری دارند زودتر اجرا می‌شوند.**  
در صورتی که دو عملگر **اولویت یکسان** داشته باشند، **از چپ به راست** اجرا می‌شوند (به جز توان `**` که از راست به چپ اجرا می‌شود).

---

## **🔹 ۱. ترتیب اولویت عملگرهای ریاضی در پایتون**
**جدول زیر ترتیب اجرای عملگرها را از بالا (بالاترین اولویت) به پایین (کمترین اولویت) نشان می‌دهد:**

| **عملگر**          | **توضیح**                 | **اولویت (از بالا به پایین)** |
|--------------------|-------------------------|------------------------------|
| `()`              | **پرانتز (اولویت بالا)** | ۱ |
| `**`             | **توان**                  | ۲ |
| `* / // %`       | **ضرب، تقسیم، تقسیم صحیح، باقی‌مانده** | ۳ |
| `+ -`            | **جمع و تفریق**           | ۴ |

📌 **نکته:** عملگر `**` (توان) از **راست به چپ** اجرا می‌شود، اما سایر عملگرها **از چپ به راست**.

---

## **🔹 ۲. مثال‌های کاربردی برای ترتیب عملیات**

</div>

```python
# بدون پرانتز، اولویت عملگرها اعمال می‌شود
result1 = 2 + 3 * 4  # ضرب اول اجرا می‌شود
print(result1)  # خروجی: 14

# استفاده از پرانتز برای تغییر اولویت
result2 = (2 + 3) * 4  # پرانتز اول اجرا می‌شود
print(result2)  # خروجی: 20

# مقایسه تقسیم و جمع
result3 = 10 + 6 / 2  # اول تقسیم اجرا می‌شود
print(result3)  # خروجی: 13.0

result4 = (10 + 6) / 2  # ابتدا جمع، سپس تقسیم
print(result4)  # خروجی: 8.0
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۳. نکات مهم درباره اولویت‌ها**
✅ **پرانتز `()`** همیشه **اولویت بالاتری دارد** و می‌توان برای کنترل ترتیب عملیات از آن استفاده کرد.  
✅ **توان `**`** از **راست به چپ** محاسبه می‌شود.  
✅ عملگرهای **ضرب `*`، تقسیم `/`، تقسیم صحیح `//` و باقی‌مانده `%` اولویت بالاتری نسبت به `+` و `-` دارند.  

---

## **🔹 ۴. مثال‌های بیشتر از ترتیب اجرای عملیات**

</div>

```python
# محاسبه توان و ضرب
print(2 ** 3 * 2)  # 2 به توان 3 → 8، سپس ضرب در 2 → خروجی: 16

# تغییر اولویت با پرانتز
print(2 ** (3 * 2))  # 3 * 2 = 6 → سپس 2 به توان 6 → خروجی: 64

# ترکیب چند عملگر
print(10 - 2 * 3 + 5)  # ضرب اول اجرا می‌شود → 10 - 6 + 5 → خروجی: 9

result = 2 ** 3 ** 2  # اجرای از راست به چپ
print(result)


# استفاده از پرانتز برای کنترل اولویت
print((10 - 2) * (3 + 5))  # ابتدا پرانتزها → 8 * 8 → خروجی: 64
```

<div dir="rtl" style="text-align: right;">

---

## **📌 نتیجه‌گیری**
✅ **پرانتز `()` بالاترین اولویت را دارد و می‌تواند ترتیب اجرا را تغییر دهد.**  
✅ **توان `**` از راست به چپ محاسبه می‌شود، اما سایر عملیات از چپ به راست انجام می‌شوند.**  
✅ **عملیات ضرب، تقسیم و باقی‌مانده اولویت بیشتری نسبت به جمع و تفریق دارند.**  
✅ **استفاده از پرانتز باعث خوانایی بهتر و جلوگیری از خطای محاسباتی می‌شود.**  

🎯 **با دانستن اولویت عملگرها، می‌توان محاسبات پیچیده را به درستی انجام داد.** 🚀
