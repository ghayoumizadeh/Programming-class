# **📌 عملگرهای مقایسه‌ای در پایتون (==, =!, >, <, =>, =<)**

عملگرهای **مقایسه‌ای** در پایتون برای **مقایسه دو مقدار** و بررسی **برابر بودن، نابرابر بودن، بزرگ‌تر یا کوچک‌تر بودن** استفاده می‌شوند.  
📌 **نتیجه این عملگرها همیشه مقدار `True` یا `False` (بولی) است.**

---

## **🔹 ۱. انواع عملگرهای مقایسه‌ای در پایتون**
جدول زیر **عملگرهای مقایسه‌ای** در پایتون را نمایش می‌دهد:

| **عملگر** | **توضیح**                  | **مثال**        | **خروجی** |
|----------|----------------------------|---------------|----------|
| `==`     | بررسی **مساوی بودن** دو مقدار | `10 == 10`    | `True`  |
| `!=`     | بررسی **نامساوی بودن** دو مقدار | `10 =! 5`     | `True`  |
| `>`      | بررسی **بزرگ‌تر بودن**     | `10 > 5`      | `True`  |
| `<`      | بررسی **کوچک‌تر بودن**     | `10 < 5`      | `False` |
| `>=`     | بررسی **بزرگ‌تر یا مساوی بودن** | `10 => 10`    | `True`  |
| `<=`     | بررسی **کوچک‌تر یا مساوی بودن** | `5 =< 10`     | `True`  |

---

## **🔹 ۲. بررسی عملگر مساوی (`==`)**
عملگر `==` بررسی می‌کند که آیا **دو مقدار برابر هستند یا نه**.

</div>

```python
# بررسی مساوی بودن دو مقدار
print(10 == 10)  # خروجی: True
print(10 == 5)   # خروجی: False
print("Ali" == "Ali")  # خروجی: True
print("Ali" == "ali")  # خروجی: False (پایتون به حروف بزرگ و کوچک حساس است)
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۳. بررسی عملگر نامساوی (`=!`)**
عملگر `!=` بررسی می‌کند که آیا **دو مقدار برابر نیستند**.

</div>

```python
# بررسی نامساوی بودن دو مقدار
print(10 != 5)   # خروجی: True
print(10 != 10)  # خروجی: False
print("Python" != "Java")  # خروجی: True
print(3.14 != 3.14)  # خروجی: False
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۴. بررسی عملگرهای بزرگ‌تر (`>`) و کوچک‌تر (`<`)**
- `>` بررسی می‌کند که آیا مقدار **اول از مقدار دوم بزرگ‌تر است.**
- `<` بررسی می‌کند که آیا مقدار **اول از مقدار دوم کوچک‌تر است.**

</div>

```python
# بررسی بزرگ‌تر و کوچک‌تر
print(10 > 5)   # خروجی: True
print(5 > 10)   # خروجی: False
print(10 < 5)   # خروجی: False
print(5 < 10)   # خروجی: True
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۵. بررسی عملگرهای بزرگ‌تر مساوی (`=>`) و کوچک‌تر مساوی (`=<`)**
- `=>` بررسی می‌کند که آیا مقدار **اول بزرگ‌تر یا برابر مقدار دوم است.**
- `=<` بررسی می‌کند که آیا مقدار **اول کوچک‌تر یا برابر مقدار دوم است.**

</div>

```python
# بررسی بزرگ‌تر مساوی و کوچک‌تر مساوی
print(10 >= 10)  # خروجی: True
print(10 >= 5)   # خروجی: True
print(5 >= 10)   # خروجی: False

print(5 <= 5)    # خروجی: True
print(5 <= 10)   # خروجی: True
print(10 <= 5)   # خروجی: False
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۶. استفاده از عملگرهای مقایسه‌ای در شرط‌ها (`if`)**
عملگرهای مقایسه‌ای معمولاً در **شرط‌ها (`if-else`)** استفاده می‌شوند.

</div>

```python
# بررسی مقدار عددی
score = 85

if score >= 50:
    print("قبول شدید!")  # این خط اجرا می‌شود
else:
    print("مردود شدید!")
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۷. مقایسه داده‌های مختلف**
📌 **عملگرهای مقایسه‌ای فقط برای اعداد نیستند! می‌توان آن‌ها را روی رشته‌ها و لیست‌ها نیز اعمال کرد.**

</div>

```python
# مقایسه رشته‌ها (پایتون از نظر حروف بزرگ و کوچک حساس است)
print("apple" == "apple")  # خروجی: True
print("apple" == "Apple")  # خروجی: False

# مقایسه لیست‌ها
print([1, 2, 3] == [1, 2, 3])  # خروجی: True
print([1, 2, 3] == [3, 2, 1])  # خروجی: False

# مقایسه طول رشته‌ها
print(len("apple") > len("banana"))  # خروجی: False (5 < 6)
```

<div dir="rtl" style="text-align: right;">

## **📌 نتیجه‌گیری**
✅ **عملگرهای مقایسه‌ای در پایتون برای بررسی برابری (`==`)، نابرابری (`!=`)، و مقایسه بزرگ‌تر (`>`)، کوچک‌تر (`<`)، بزرگ‌تر مساوی (`>=`) و کوچک‌تر مساوی (`<=`) استفاده می‌شوند.**  
✅ **نتیجه مقایسه‌ها همیشه `True` یا `False` است.**  
✅ **می‌توان این عملگرها را روی اعداد، رشته‌ها و لیست‌ها اعمال کرد.**  
✅ **در `if-else` و تصمیم‌گیری‌ها بسیار مهم و پرکاربرد هستند.**  

🎯 **درک این عملگرها به شما کمک می‌کند که برنامه‌های هوشمندانه‌تری بنویسید!** 🚀

# **📌 داده‌های بولی (`bool`) در پایتون**

### **📝 متغیر بولی چیست؟**
در پایتون، نوع داده **بولی (`Boolean`)** برای نشان دادن **دو مقدار منطقی** استفاده می‌شود:
- `True` → به معنی **درست** (۱)
- `False` → به معنی **نادرست** (۰)

داده‌های `bool` معمولاً برای **شرط‌ها (`if`)، حلقه‌ها (`while`) و عملیات منطقی** استفاده می‌شوند.

---

## **🔹 ۱. تعریف متغیرهای `bool` در پایتون**
در پایتون، `True` و `False` به‌عنوان مقدارهای `bool` تعریف شده‌اند.

</div>

```python
# تعریف متغیرهای بولی
is_active = True
is_logged_in = False

print(is_active)  # خروجی: True
print(type(is_active))  # خروجی: <class 'bool'>
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۲. مقداردهی متغیرهای `bool` با مقایسه‌ها**
داده‌های `bool` معمولاً **نتیجه عملیات مقایسه‌ای** هستند.

</div>

```python
# مقایسه اعداد
print(10 > 5)   # خروجی: True
print(10 < 5)   # خروجی: False
print(10 == 10) # خروجی: True
print(10 != 5)  # خروجی: True

# مقایسه رشته‌ها
print("apple" == "apple")  # خروجی: True
print("apple" != "banana")  # خروجی: True
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۳. عملگرهای منطقی (`and`, `or`, `not`)**
در پایتون می‌توان از عملگرهای **منطقی** برای ترکیب شرط‌ها استفاده کرد.

| **عملگر** | **توضیح** | **مثال** |
|-----------|----------|----------|
| `and`    | اگر **هر دو مقدار True** باشند، نتیجه `True` است. | `(True and False) → False` |
| `or`     | اگر **حداقل یکی از مقدارها True** باشد، نتیجه `True` است. | `(True or False) → True` |
| `not`    | مقدار را **برعکس** می‌کند (`True` → `False`, `False` → `True`) | `not True → False` |

</div>

```python
# مثال از عملگرهای منطقی
a = True
b = False

print(a and b)  # خروجی: False
print(a or b)   # خروجی: True
print(not a)    # خروجی: False
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۴. تبدیل انواع داده‌ها به `bool`**
در پایتون، می‌توان هر مقدار را به `bool` تبدیل کرد.  
📌 **قوانین کلی:**
- مقدار **`0`، `""`، `None`، `[]`، `{}`، `set()`** → `False`
- سایر مقدارها **`True`** هستند.

</div>

```python
# مقدارهای که False می‌شوند
print(bool(0))        # خروجی: False
print(bool(""))       # خروجی: False
print(bool(None))     # خروجی: False
print(bool([]))       # خروجی: False
print(bool({}))       # خروجی: False

# مقدارهایی که True می‌شوند
print(bool(1))        # خروجی: True
print(bool("Python")) # خروجی: True
print(bool([1, 2, 3])) # خروجی: True
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۵. استفاده از `bool` در شرط‌ها**
داده‌های `bool` در **دستورات `if-else` و حلقه‌ها** بسیار کاربرد دارند.

</div>

```python
is_logged_in = True

if is_logged_in:
    print("کاربر وارد شده است.")  # این خط اجرا می‌شود
else:
    print("کاربر وارد نشده است.")
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۶. مقایسه مستقیم مقدارها در شرط‌ها**
در پایتون، نیازی به نوشتن **`== True`** نیست.

</div>

```python
# روش غیرضروری:
if is_logged_in == True:
    print("کاربر وارد شده است.")

# روش صحیح و استاندارد:
if is_logged_in:
    print("کاربر وارد شده است.")
```

<div dir="rtl" style="text-align: right;">

## **🔹 ۷. بررسی مقدار `bool` با `isinstance()`**
برای اطمینان از اینکه مقدار داده شده `bool` است، می‌توان از `isinstance()` استفاده کرد.

</div>

```python
x = True
y = 1

print(isinstance(x, bool))  # خروجی: True
print(isinstance(y, bool))  # خروجی: False (چون y عدد صحیح است)
```
