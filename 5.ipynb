# **📌 دستور شرطی `if` در پایتون**

### **📝 دستور `if` چیست؟**
در پایتون، از **`if`** برای **اجرای کدها به‌صورت شرطی** استفاده می‌شود.  
**اگر شرط برقرار باشد (`True`)، کد داخل `if` اجرا می‌شود، در غیر این صورت (`False`) اجرا نمی‌شود.**  

---

## **🔹 ۱. ساختار کلی `if` در پایتون**
📌 **دستور `if`** همیشه با یک **شرط** همراه است.  
🔹 **اگر شرط برقرار باشد (`True`)، کدهای داخل `if` اجرا می‌شوند.**  
🔹 **اگر شرط برقرار نباشد (`False`)، کدهای `if` اجرا نمی‌شوند.**  

</div>

```python
# ساختار کلی if
if شرط:
    # بلوک کد اجرا می‌شود اگر شرط True باشد
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۲. مثال ساده از `if`**
📌 در این مثال، اگر مقدار عددی **بزرگ‌تر از ۱۰** باشد، پیام `"عدد بزرگ‌تر از ۱۰ است!"` نمایش داده می‌شود.

</div>

```python
number = 15

if number > 10:
    print("عدد بزرگ‌تر از ۱۰ است!")  # این خط اجرا می‌شود
```

<div dir="rtl" style="text-align: right;">

📌 **اگر مقدار عددی ۵ باشد، شرط برقرار نخواهد بود و هیچ پیامی نمایش داده نمی‌شود.**

</div>

```python
number = 5

if number > 10:
    print("عدد بزرگ‌تر از ۱۰ است!")  # این خط اجرا نمی‌شود
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۳. دستور `if-else`**
📌 **اگر شرط برقرار نباشد (`False`)، قسمت `else` اجرا می‌شود.**  
🔹 از `else` برای **اجرای کد جایگزین در صورت برقرار نبودن شرط** استفاده می‌شود.

</div>

```python
number = 8

if number > 10:
    print("عدد بزرگ‌تر از ۱۰ است!")
else:
    print("عدد ۱۰ یا کوچک‌تر است.")  # این خط اجرا می‌شود
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۴. استفاده از `if-elif-else`**
📌 **اگر چندین شرط داشته باشیم، از `elif` استفاده می‌کنیم.**  
🔹 `elif` مخفف **"else if"** است و برای بررسی شرایط دیگر به کار می‌رود.  
🔹 اگر **هیچ‌یک از `if` یا `elif`ها اجرا نشود، `else` اجرا می‌شود.**  

</div>

```python
score = 75

if score >= 90:
    print("نمره عالی!")
elif score >= 70:
    print("نمره خوب!")  # این خط اجرا می‌شود
elif score >= 50:
    print("نیاز به تلاش بیشتر!")
else:
    print("مردود شدید!")
```

<div dir="rtl" style="text-align: right;">

📌 **در این مثال، مقدار `score` برابر `75` است، بنابراین `elif score >= 70` اجرا شده و `"نمره خوب!"` نمایش داده می‌شود.**

---

## **🔹 ۵. استفاده از `if` با عملگرهای منطقی (`and`, `or`, `not`)**
📌 می‌توان از **`and`, `or`, `not`** برای ترکیب شرط‌ها استفاده کرد.  

</div>

```python
age = 20
has_id = True

# استفاده از and (هر دو شرط باید True باشند)
if age >= 18 and has_id:
    print("شما می‌توانید وارد شوید!")  # این خط اجرا می‌شود

# استفاده از or (حداقل یکی از شرط‌ها باید True باشد)
if age >= 18 or has_id:
    print("احتمالاً ورود مجاز است.")  # این خط اجرا می‌شود

# استفاده از not (برعکس مقدار منطقی)
if not has_id:
    print("لطفاً کارت شناسایی خود را ارائه دهید.")
```

<div dir="rtl" style="text-align: right;">

---

## **🔹 ۶. `if` در مقایسه مقادیر و داده‌های متنی**
📌 **`if` را می‌توان برای مقایسه مقادیر عددی، رشته‌ای، و حتی لیست‌ها به کار برد.**

</div>

```python
# مقایسه رشته‌ها (پایتون به حروف بزرگ و کوچک حساس است)
fruit = "Apple"

if fruit == "apple":
    print("میوه انتخاب شده: سیب")  # این شرط False است و اجرا نمی‌شود
elif fruit == "Apple":
    print("میوه انتخاب شده: سیب (حروف بزرگ)")  # این خط اجرا می‌شود
```

<div dir="rtl" style="text-align: right;">

📌 **توجه:** پایتون به **حروف بزرگ و کوچک حساس است** (`"Apple" != "apple"`).

---

## **🔹 ۷. مقداردهی شرطی در یک خط (`if` کوتاه)**
📌 **می‌توان از `if` در یک خط برای مقداردهی استفاده کرد.**  

</div>

```python
age = 20
status = "بزرگسال" if age >= 18 else "نوجوان"

print(status)  # خروجی: بزرگسال
```

<div dir="rtl" style="text-align: right;">

📌 در این مثال، اگر `age` بزرگ‌تر یا مساوی `18` باشد، مقدار `"بزرگسال"` در `status` ذخیره می‌شود، در غیر این صورت مقدار `"نوجوان"` ذخیره خواهد شد.

---

## **🔹 ۸. بررسی مقدارهای خاص در `if`**
📌 **در پایتون، مقدارهایی مثل `0`, `None`, `False`, `""` (رشته خالی) همگی `False` محسوب می‌شوند.**

</div>

```python
value = ""

if value:
    print("مقدار معتبر است.")
else:
    print("مقدار نامعتبر است.")  # این خط اجرا می‌شود
```

<div dir="rtl" style="text-align: right;">

📌 **چون `value` مقدار `""` (رشته خالی) دارد، شرط `if` برقرار نمی‌شود و `else` اجرا می‌شود.**

✅ **`if-elif-else` برای بررسی چندین شرط به کار می‌رود.**  
✅ **عملگرهای منطقی `and`, `or`, `not` برای ترکیب شرط‌ها استفاده می‌شوند.**
